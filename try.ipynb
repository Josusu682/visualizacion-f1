```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Cargar los JSON originales
ruta_v1 = '2025/Australian Grand Prix/Qualifying/PIA/1_tel.json'
ruta_v5 = '2025/Australian Grand Prix/Qualifying/PIA/5_tel.json'

df_v1_raw = pd.read_json(ruta_v1)
df_v5_raw = pd.read_json(ruta_v5)

# 2. EL TRUCO: Transformar la columna 'tel' (que contiene listas) en un DataFrame plano
df_v1 = pd.DataFrame(df_v1_raw['tel'].to_dict())
df_v5 = pd.DataFrame(df_v5_raw['tel'].to_dict())

# (Opcional) Si haces un display ahora, verás que las columnas por fin son time, rpm, speed...
display(df_v5.head())

# 3. ¡A graficar! Usamos 'time' en el eje X y 'speed' en el eje Y (todo en minúsculas)
plt.figure(figsize=(15, 5))

# Superponer ambas líneas de velocidad
plt.plot(df_v1['time'], df_v1['speed'], label='Vuelta 1 (Q1)', color='gray', alpha=0.7)
plt.plot(df_v5['time'], df_v5['speed'], label='Vuelta 5 (Q3)', color='orange', linewidth=2)

plt.title('Comparación de Velocidad por Tiempo: Piastri (V1 vs V5) - Australia 2025')
plt.xlabel('Tiempo de vuelta (segundos)')
plt.ylabel('Velocidad (km/h)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
